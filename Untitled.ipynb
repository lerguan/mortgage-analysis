```python
df.drop(columns=['id', 'member_id', 'url', 'desc', 'emp_title', 'sec_app_earliest_cr_line', 'sec_app_inq_last_6mths', 
                 'sec_app_mths_since_last_major_derog'], inplace=True)
```

```python
# identify percentage of null values in each column
df_null_value = df.isnull().sum()/df.shape[0]
```

```python
# identify the columns that contains more than 80% missing values
columns_null_80percent = list(df_null_value.loc[df_null_value >= 0.8].index)
```

```python
# drop columns with null values more than 80%
df.drop(columns=columns_null_80percent, inplace=True)
```

```python
df.loan_status.value_counts()
```

```python
def badloan(row):
    if row['loan_status']=='Current' or row['loan_status']=='Issued' or row['loan_status']=='Fully Paid':
        return 0
    elif row['loan_status']=='In Grace Period' or row['loan_status']=='Late (31-120 days)' or row['loan_status']=='Late (16-30 days)':
        return 1
```

```python
# add new column `loan_late` indicates bad loan
df['loan_late']=df.apply(lambda row: badloan(row), axis=1)
```

```python
df.emp_length_yr.value_counts()
```

```python
df.groupby('emp_length_yr').loan_late.mean().plot(kind='bar', ylabel='Average number of bad loans', \
                                                  title='The Average number of bad loans by the length of employment');
```

```python
plt.scatter(x=df.annual_inc, y=df.loan_late)
plt.xlim(-100000, 1000000)
```

```python
plt.scatter(x=df.groupby('annual_inc').count().index, y=df.groupby('annual_inc').loan_late.sum())
plt.xlim(-100000, 1000000)
plt.xlabel('Annual Income')
plt.ylabel('Number of Late Loans')
plt.title('The Number of Late Loans by Annual Income');
```
